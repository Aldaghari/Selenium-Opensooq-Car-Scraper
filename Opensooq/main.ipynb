```python
# Test
from selenium import webdriver
from selenium.webdriver.remote import webelement
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from Car import Car
driver = webdriver.Chrome()
def get_car_info(link: str) -> dict:

    website_url: str = link

    cars = list[Car]
    
    driver.get(website_url)
    
    car_dict: dict[str, str] = {
        "Price" : "0.0 OMR"
    }

    car_price: str = driver.find_element(By.XPATH, "//section[@id='postViewCardDesktop']/div[1]").text
    car_dict["Price"] = car_price

    info_table = driver.find_element(By.XPATH, "//section[@id='PostViewInformation']/ul[1]")

    for info in info_table.find_elements(By.TAG_NAME, "li"):
        x: list = info.text.split("\n", 1)
        car_dict[x[0]] = x[1]


    print(car_dict)
    return car_dict


def select_car_post(start_url):
    cars: list[Car] = []
    cars_table = driver.find_element(By.ID, "serpMainContent")
    car_dict: dict[str, str]
    links: list[str] = []
    # Get all the links
    for car in cars_table.find_elements(By.XPATH, "div"):
        if car.get_attribute("data-id") != None:
            links.append(car.find_element(By.TAG_NAME, "a").get_attribute("href"))
            
    # Retrive car info       
    for link in links:
        driver.get(link)
        car_dict = get_car_info(link)
        cars.append(Car(car_dict))
        driver.get(start_url)

    # Go to the next page

# A bit of hard coding, consider changing later
def select_sort(url) -> str:
    # Lazy implementation, use click() function instead
    sort_options: list[(str, str)] = [
        ("Relevant" , ""),
        ("Date: Newest", "?search=true&sort_code=recent"),
        ("Price: Ascending", "?search=true&sort_code=price-"),
        ("Price: Descending", "?search=true&sort_code=price+"), 
        ("Year: Old First", "?search=true&sort_code=Car_Year-"),
        ("Year: New First", "?search=true&sort_code=Car_Year+"),
        ("Kilometers: Lowest First", "?search=true&sort_code=Kilometers_Cars-"),
        ("Kilometers: Highest First", "?search=true&sort_code=Kilometers_Cars+")
    ]
    print("Select The Search Sort:")
    count: int = 1
    for o in sort_options:
        print(str(count) + ".", o[0])
        count += 1
     

    # inside the array range
    selection: int = int(input("Enter a number: ")) - 1

    driver.get(url + sort_options[selection][1])

    return sort_options[selection][1]

def select_car_brand() -> str:
    brand_of_cars_url: str = "https://om.opensooq.com/en/cars"
    driver.get(brand_of_cars_url)
    car_brand = driver.find_element(By.CLASS_NAME, "gridCont")
    car_brand_names_and_links: list[(str, str)] = []

    for type in car_brand.find_elements(By.TAG_NAME, "a"):
        car_name: str = " ".join(type.find_element(By.TAG_NAME, "div").text.split())
        car_link: str = type.get_attribute("href")
        car_brand_names_and_links.append((car_name, car_link))


    print("Choose a Car Brand:")
    count: int = 1
    for c in car_brand_names_and_links:
        print(str(count) + ".",  c[0])
        count += 1
    print()
    # inside the array range
    selection: int = int(input("Enter a number: ")) - 1

    driver.get(car_brand_names_and_links[selection][1])

    return car_brand_names_and_links[selection][1]


def main():
    brand_url: str = select_car_brand()
    sort: str = select_sort(brand_url)
    post_url: str = brand_url + sort
    select_car_post(post_url)
    driver.close()

main()
```
